```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load the CSV files into DataFrames
df1 = pd.read_csv('dataset/lower_income1.csv')
df2 = pd.read_csv('dataset/upper_income1.csv')
```

```python
print("Lower income csv columns:", df1.columns)
print("Upper income csv columns:", df2.columns)
```

```python
# Set plot style
sns.set_theme(style="whitegrid")

# 1. Category Distribution (Count of each category) - Bar Chart
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.countplot(data=df1, x='Category', order=df1['Category'].value_counts().index, palette='Blues')
plt.title('Category Distribution - Lower Income')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
sns.countplot(data=df2, x='Category', order=df2['Category'].value_counts().index, palette='Oranges')
plt.title('Category Distribution - Upper Income')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# 2. Average Amount per Category - Bar Chart
avg_lower = df1.groupby('Category')['Amount'].mean().sort_values(ascending=False)
avg_upper = df2.groupby('Category')['Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
avg_lower.plot(kind='bar', color='lightblue', width=0.8)
plt.title('Average Amount per Category - Lower Income')
plt.xlabel('Category')
plt.ylabel('Average Amount')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
avg_upper.plot(kind='bar', color='salmon', width=0.8)
plt.title('Average Amount per Category - Upper Income')
plt.xlabel('Category')
plt.ylabel('Average Amount')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# 3. Income vs Expense - Stacked Bar Chart
lower_income_expense = df1.groupby(['Category', 'Type']).size().unstack(fill_value=0)
upper_income_expense = df2.groupby(['Category', 'Type']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
lower_income_expense.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], ax=plt.gca())
plt.title('Income vs Expense - Lower Income')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
upper_income_expense.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], ax=plt.gca())
plt.title('Income vs Expense - Upper Income')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
